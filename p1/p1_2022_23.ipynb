Remember that [Numpy](https://docs.scipy.org/doc/numpy-1.13.0/reference/) allows us to define images of different kinds treating them as matrices.

The most common examples are:

* Binary: with values of 0 or 1, interpreted as black or white, respectively.
* Grayscale: with possible data types uint8, uint16, double, whose pixel values define their intensity.
* RGB Image: color images consisting of 3 matrices (usually called 'channels'), which contain intensity values for each color separately R (red), G (green) and B (blue).

a) Read the image `images_notebook/circles.bmp` (see figure top) and define 3 images in such a way that each one contains only 1 ball (see figure bottom). Visualize the 3 images in the same figure, each one with its corresponding title.

<img src="images_notebook/ex16.png">

**Note:** Be careful when applying `imshow` since it automatically increases the contrast of the image (putting minimum values to 0 and maximal to 1).

**Note:** Be careful when applying type conversion. Check the ranges of the image values.

```python
#Your solution
# Load the image.
circles = io.imread('images_notebook/circles.bmp')
# Convert the image in grayscale.
circles_gray = rgb2gray(circles)
plt.imshow(circles_gray, cmap='gray')

# Get all the colors that appear in the image.
colors = np.unique(circles_gray) 

#color = [0. , 0.49803922 , 0.76470588 , 1. ] 0 is black, 1 is white

#in order to show black dot, we create a copy of the original image
circle_black = circles_gray.copy()
#then we set as 1 wich corresponds to color white, 
# all colors diferent from colors[0]=0. wich corresponds to black
circle_black[circles_gray!=colors[0]]=1

#in order to show dark gray dot, we create a copy of the original image
circle_dark_gray = circles_gray.copy()
#then we set as 1 wich corresponds to color white, 
# all colors diferent from colors[1]=0.49803922 wich corresponds to dark gray
circle_dark_gray[circles_gray!=colors[1]]=1

#in order to show light gray dot, we create a copy of the original image
circle_light_gray = circles_gray.copy()
#then we set as 1 wich corresponds to color white, 
# all colors diferent from colors[2]=0.76470588 wich corresponds to light gray
circle_light_gray[circles_gray!=colors[2]]=1

#showing all 3 images with the title of each one
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(20,20))

ax = axs[0]
ax.imshow(circle_black, cmap='gray', vmin = 0, vmax = 1)
ax.axis('off')
ax.set_title('Black circle')

ax = axs[1]
ax.imshow(circle_dark_gray, cmap='gray',  vmin = 0, vmax = 1)
ax.axis('off')
ax.set_title('Dark gray circle')

ax = axs[2]
ax.imshow(circle_light_gray, cmap='gray', vmin = 0, vmax = 1)
ax.axis('off')
ax.set_title('Light gray circle')

plt.show()
```

b) Given the images `images_notebook/coat.png` and `images_notebook/model.png` (first two images on the figure below), implement the necessary code to construct the image of the model with the coat (third image). Next find in internet an image with a texture and apply it to obtain an image of the jacket with the found texture (fourth image is only an example).

<img src="images_notebook/ex17d.png">

**Note:** when applying [resize, rotate, etc...](https://scikit-image.org/docs/stable/auto_examples/transform/plot_rescale.html) keep track how the image type and values range change. When necessary, apply type conversion using [img_as_float, img_as_uint, img_as_ubyte.](https://scikit-image.org/docs/stable/api/skimage.html)

**Note: DO NOT USE LOOPS to overlap the images!! It will be penalized.**


```python
#Your solution
#some needed imports
from skimage.util import img_as_ubyte
from skimage.color import rgba2rgb

#reading images
coat = io.imread('images_notebook/coat.png')
model = io.imread('images_notebook/model.png')

#now i make a copy of the model to work with the copy and not making changes the original one
model_copy = model.copy()
#i set as white pixels wich coincide with the pixels different from black in the coat image
model_copy[coat!=0] = 255
#so now i can merge the two images without unexpected color changes
merged_img = rgba2rgb(model_copy + coat)

#showing all images
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,15))

ax = axs[0]
ax.imshow(model)
ax.axis('off')
ax.set_title('Model Image')

ax = axs[1]
ax.imshow(coat)
ax.axis('off')
ax.set_title('Coat Image')

ax = axs[2]
ax.imshow(merged_img)
ax.axis('off')
ax.set_title('Merged Image')

plt.show()

#and the info
print ("\nMerged Image", "\nDtype: ", merged_img.dtype, "\nShape: ", merged_img.shape, "\nMin: ", merged_img.min(), "\nMax: ", merged_img.max()) 
```
